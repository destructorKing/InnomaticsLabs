```python
#Social Media – Post Engagement Analyzer
def analyze_engagement(likes):
    total_likes = 0
    for like in likes:
        total_likes += like

    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"

    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

analyze_engagement([200, 300, 150, 400])
```

```python
#Healthcare – Medicine Stock Alert System
def check_medicine_stock(stock):
    status = "Low Stock Alert" if stock < 10 else "Stock Sufficient"

    print(f"Medicine Stock: {stock}")
    print(f"Status: {status}")

check_medicine_stock(6)
```

```python
#Agriculture – Rainfall Adequacy Checker
def rainfall_checker(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    status = "Adequate Rainfall" if average >= required_level else "Inadequate Rainfall"

    print(f"Average Rainfall: {average}")
    print(f"Rainfall Status: {status}")


rainfall_checker([60, 80, 70, 78], 70)
```

```python
#Social Media – Duplicate Account Detection
def detect_duplicates(usernames):
    unique_users = set(usernames)

    status = "Yes" if len(unique_users) != len(usernames) else "No"

    print(f"Duplicate Accounts Found: {status}")
    
detect_duplicates(["abhi", "rahul", "abhi", "neha"])
```

```python
#Agriculture – Premium Crop Price Filter
def filter_premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print(f"Premium Crops: {premium}")


filter_premium_crops([1500, 2500, 1800, 3200])
```

```python
#System Monitoring – Application Health Checker
def check_system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {errors}")
    print(f"System Status: {status}")


errors=int(input("Enter the number of errors detected: "))
check_system_health(errors)
```

```python
#Banking – Daily Transaction Limit Checker
def transaction_limit_checker(amount):
    status = "Approved" if amount <= 50000 else "Rejected"

    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")

transaction_limit_checker(60000)
```

```python
#E-Learning – Student Attendance Eligibility System
def attendance_eligibility(attended_classes, total_classes):
    percentage = (attended_classes / total_classes) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"

    print(f"Attendance Percentage: {percentage}")
    print(f"Exam Eligibility: {status}")

attended=int(input("Enter the number of classes attended: "))
total=int(input("Enter the total number of classes: "))
attendance_eligibility(attended, total)
```

```python
#Smart Electricity Bill Analyzer
def calculate_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

bill, usage = calculate_bill(250)
print(f"Total Bill: ₹{bill}")
print(f"Usage Classification: {usage}")
```
