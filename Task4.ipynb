```python
#Mobile Recharge Validation
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Minimum recharge is ₹50")
        return False

    if amount not in valid_plans:
        print("Invalid plan. Valid plans:", valid_plans)
        return False

    print("Recharge successful!")
    return True


while True:
    amt = int(input("Enter recharge amount: ₹"))
    if validate_recharge(amt):
        break
    print("Please try again...\n")
```

```python
#nventory Reorder Alert System
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}:Reorder Alert (Stock = {stock})")
        else:
            print(f"{product}:Stock OK (Stock = {stock})")


inventory = {
    "Keyboard": 12,
    "Mouse": 25,
    "Monitor": 8,
    "USB Cable": 40
}

check_inventory(inventory)
```

```python
#Student Result Processing System
def process_result(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


student_marks = [60, 45, 70, 55]
result = process_result(student_marks)
print("Result:", result)
```

```python
#Cab Fare Estimator
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)

    if peak:
        fare *= 1.25  # 25% extra

    return fare


while True:
    try:
        km = float(input("Enter distance (km): "))
        peak_input = input("Peak hour? (yes/no): ").lower()

        peak_hour = peak_input == "yes"

        fare = calculate_fare(km, peak_hour)
        print(f"Estimated Fare: ₹{fare:.2f}")
        break

    except ValueError:
        print("Invalid input. Please retry.\n")
```

```python
#Attendance Eligibility Checker
def check_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    print("Attendance %:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


record = ["P", "A", "P", "P", "A", "P", "P", "P"]
status = check_eligibility(record)
print("Status:", status)
```

```python
#Password Strength Checker
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak: Minimum 8 characters required"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if not has_digit:
        return "Weak: Must contain a digit"

    if not has_special:
        return "Weak: Must contain special character (@#$)"

    return "Strong Password"


pwd = input("Enter password: ")
print(check_password(pwd))
```
