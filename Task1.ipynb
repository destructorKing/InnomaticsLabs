```python
#User Login Check
username = "admin"
password = "1234"

user=input("Enter Username:")
passw=input("Enter Password:")

if user == username and passw == password:
    print("Login Successful!")
else:
    print("Invalid Credentials!")
```

```python
#Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
passCount=0
failCount=0

for i in marks:
    if i>=50:
        passCount +=1
    else:
        failCount +=1
print("Pass Students:", passCount)
print("Fail Students:", failCount)
```

```python
#Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]

new_names = []

for name in names:
    new_names.append(name.strip().lower())

print("Cleaned Names:", new_names)
```

```python
#Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]

for mssg in messages:
    length = len(mssg)
    print(f"Message: '{mssg}' ; Length: {length}")
    
    if length > 10:
        print("Long message detected")
```

```python
#Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR")

print("Total ERROR entries:", error_count)
```
