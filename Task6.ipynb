```python
#Smart Parking Lot Management System
def smart_parking_system(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT" and current > 0:
            current -= 1

        peak = max(peak, current)

    print(f"Currently Parked Vehicles: {current}")

    if current >= capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")

    print(f"Peak Parking Usage: {peak}")

smart_parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
#Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5  
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 5

    if weather == "Rainy":
        base_time += 10
    elif weather == "Stormy":
        base_time += 20

    print(f"Estimated Delivery Time: {base_time} minutes")

estimate_delivery_time(8, "High", "Rainy")
```

```python
#Movie Theatre Seat Occupancy Analyzer
def seat_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print(f"Occupancy: {int(occupancy)}%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")

seat_occupancy(200, [1] * 150)
```

```python
#Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    avg_cpu = sum(cpu_readings) / len(cpu_readings)

    print(f"Average CPU Load: {int(avg_cpu)}%")

    if avg_cpu < 50:
        print("Server Status: Normal")
    elif 50 <= avg_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

classify_server_load([45, 60, 70, 85, 90])
```

```python
#Smart Classroom Resource Usage Monitor
def classroom_monitor(usage):
    overused = []

    for resource, hours in usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

classroom_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
#Online Event Registration Capacity Controller
def registration_controller(capacity, registrations):
    if registrations > capacity:
        waitlisted = registrations - capacity
        confirmed = capacity
    else:
        confirmed = registrations
        waitlisted = 0

    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")

    if waitlisted > 0:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")

registration_controller(100, 105)
```
