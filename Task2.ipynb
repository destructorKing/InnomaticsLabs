```python
#Unique Words in a Sentence
sentence = input("Enter sentence: ")

words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
#Highest Salary from Employee Data
employees = {
    "Alex": 45000,
    "Claire": 67000,
    "Nancy": 75000
}

highest_paid = max(employees, key=employees.get)

print("Highest Salary:", highest_paid, "-", employees[highest_paid])
```

```python
#Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]

max_val = numbers[0]
min_val = numbers[0]

for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

```python
#Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
#Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
percentage = (present_count / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

```python
#Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

```python
#Count Character Frequency
text = "Innomatics Research Labs"

freq = {}

for char in text:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

print(freq)
```

```python
#Check if a Key Exists in a Dictionary
employees = {"Alex": 75000, "Claire": 68000}

key = input("Enter employee name to check: ")

if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```
